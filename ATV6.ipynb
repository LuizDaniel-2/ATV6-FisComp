### a)

### c)
Crie um gráfico mostrando como o erro relativo na subtração $(1+x)−1$ se comporta para valores de $x$ entre $10^{(−1)}$ e $10^{(−16)}$. Use `numpy` e `matplotlib`.

O que acontece quando $x$ se aproxima de zero?

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Geração dos dados de X
# Usamos np.logspace para criar valores de x uniformemente espaçados em uma escala logarítmica.
# Isso é essencial para visualizar o comportamento em várias ordens de magnitude, de 10^-1 a 10^-16.
x = np.logspace(-16, -1, num=1000, dtype=np.float64)

# 2. Cálculo da Expressão e do Erro Relativo
# Valor Exato: Matematicamente, (1 + x) - 1 = x
valor_exato = x

# Valor Aproximado: Este é o resultado que o computador calcula usando a aritmética de ponto flutuante padrão.
# A imprecisão acontece aqui, especialmente para x pequeno.
valor_calculado = (1.0 + x) - 1.0

# Cálculo do Erro Relativo: |Valor Calculado - Valor Exato| / |Valor Exato|
# O erro é a medida de quão longe o resultado do computador está do resultado real.
erro_relativo = np.abs(valor_calculado - valor_exato) / np.abs(valor_exato)

# 3. Criação do Gráfico
plt.figure(figsize=(12, 7))

# Plotamos o erro relativo em função de x.
plt.plot(x, erro_relativo, 'o-', markersize=3, label='Erro Relativo')

# Usamos escalas logarítmicas para ambos os eixos para visualizar claramente a variação.
plt.xscale('log')
plt.yscale('log')

# Adicionamos títulos e rótulos para clareza
plt.title('Erro Relativo na Computação de f(x) = (1 + x) - 1', fontsize=16)
plt.xlabel('Valor de x (Escala Logarítmica)', fontsize=12)
plt.ylabel('Erro Relativo (Escala Logarítmica)', fontsize=12)

# Adiciona uma linha para marcar o "Épsilon da Máquina", o limite da precisão do float.
plt.axvline(x=np.finfo(float).eps, color='r', linestyle='--', label=f'Épsilon da Máquina (~{np.finfo(float).eps:.1e})')
plt.legend()

# Exibe o gráfico
plt.show()
```

# d) (Exploração mais desafiadora)

Implemente uma função que calcule numericamente a derivada de $f(x)=ln(x)$ no ponto $x=1$ , usando a fórmula da diferença finita progressiva:

$$f'(x)≈\frac{f(x+h)−f(x)}{h}$$

- Calcule o erro absoluto em relação ao valor exato da derivada $f'(1)=1$ para valores de $h=10^{−}1,10^{−2},\dots,10^{−16}$

- Plote o erro absoluto em função de $h$ em escala $\log-\log$
- Identifique o intervalo de $h$ em que o erro é minimizado e explique o motivo do aumento do erro para valores de $h$ muito pequenos.

Reflexão:
Mesmo operações simples, como somar e subtrair, podem introduzir erros quando feitas com ponto flutuante. Este exercício mostra como números muito pequenos podem ser "perdidos" em meio a números grandes e como isso impacta a acuracidade dos cálculos. Além disso, aproximações numéricas como derivadas finitas devem equilibrar erro de truncamento com erro de arredondamento para obter resultados confiáveis.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definição da função f(x) = ln(x)
def f(x):
    """Calcula o logaritmo natural de x."""
    return np.log(x)

# Definição da derivada numérica pela fórmula da diferença finita progressiva
def derivada_progressiva(f, x, h):
    """Calcula a derivada de f(x) no ponto x com um passo h."""
    return (f(x + h) - f(x)) / h

# --- Parâmetros do Problema ---
x_ponto = 1  # Ponto onde a derivada será calculada

# O valor exato da derivada de ln(x) em x=1 é 1/x = 1/1 = 1.
valor_exato = 1.0

# Cria um array de valores para h, de 10^-16 até 10^-1.
# Usamos dtype=np.float64 para garantir a precisão dupla.
h_valores = np.logspace(-16, -1, num=10000, dtype=np.float64)

# --- Cálculos ---
# Calcula a derivada numérica para cada valor de h
derivada_calculada = derivada_progressiva(f, x_ponto, h_valores)

# Calcula o erro absoluto para cada valor de h
erro_absoluto = np.abs(derivada_calculada - valor_exato)

# --- Geração do Gráfico ---
plt.figure(figsize=(12, 8))

# Plota h vs. erro absoluto em escala log-log
plt.loglog(h_valores, erro_absoluto, '-', markerfacecolor='royalblue', label='Erro Absoluto')

# Encontra o ponto de erro mínimo para destacar no gráfico
min_error_idx = np.argmin(erro_absoluto)
h_otimo = h_valores[min_error_idx]
min_erro = erro_absoluto[min_error_idx]
plt.loglog(h_otimo, min_erro, 'r*', markersize=15, label=f'Erro Mínimo em h ≈ {h_otimo:.1e}')

# Configurações do gráfico
plt.title('Erro na Derivada Numérica de ln(x) em x=1', fontsize=16)
plt.xlabel('Tamanho do Passo (h)', fontsize=12)
plt.ylabel('Erro Absoluto', fontsize=12)
plt.gca().invert_xaxis() # Inverte o eixo x para ler da direita para a esquerda
plt.legend()
plt.grid(False)
plt.show()

print(f"O erro é minimizado para h ≈ {h_otimo:.2e}")
print(f"O valor do erro mínimo é ≈ {min_erro:.2e}")
```

### a)

O problema: multiplicação escalar-vetor

Considere o problema de multiplicar cada elemento de um vetor por um número escalar $\alpha$:

$y_i=\alpha\,x_i$ para $i=1,2,\dots,N$

Implemente isso em Python, de pelo menos duas maneiras:

1. Usando um laço `for` puro.
2. Usando uma abordagem vetorizada com `numpy`.

```python
import numpy as np
import time

# --- 1. Configuração do Problema ---

# Define o tamanho do vetor. Usamos um número grande para que a diferença de tempo seja significativa.
N = 10**8  # 10 milhões de elementos
# Define o escalar a ser multiplicado
alpha = 2.5

# Cria os dados de entrada
# Um vetor como lista pura do Python para o laço 'for'
x_lista = list(range(N))
# Um vetor NumPy para a abordagem vetorizada
x_numpy = np.arange(N)


# --- 2. Abordagem 1: Laço 'for' puro ---

print(f"Iniciando cálculo com Laço 'for' puro para N = {N}...")

# Inicia a contagem do tempo
start_time_loop = time.time()

# Inicializa a lista de resultado
y_lista = [0] * N # Pré-alocar a lista tem um desempenho um pouco melhor que .append()

# Itera sobre cada elemento, realiza a multiplicação e armazena na nova lista
for i in range(N):
    y_lista[i] = alpha * x_lista[i]

# Finaliza a contagem do tempo
end_time_loop = time.time()
tempo_loop = end_time_loop - start_time_loop
print(f"Cálculo com Laço 'for' concluído.")
print(f"Tempo de execução: {tempo_loop:.4f} segundos\n")


# --- 3. Abordagem 2: Vetorização com NumPy ---

print(f"Iniciando cálculo com NumPy vetorizado para N = {N}...")

# Inicia a contagem do tempo
start_time_numpy = time.time()

# A operação é realizada em todos os elementos do vetor de uma só vez.
y_numpy = alpha * x_numpy

# Finaliza a contagem do tempo
end_time_numpy = time.time()
tempo_numpy = end_time_numpy - start_time_numpy
print(f"Cálculo com NumPy concluído.")
print(f"Tempo de execução: {tempo_numpy:.4f} segundos\n")


# --- 4. Análise e Comparação ---

print("--- Comparação ---")
print(f"A abordagem com NumPy foi aproximadamente {tempo_loop / tempo_numpy:.3f} vezes mais rápida.")
```

# Exercício 3

A função seno pode ser aproximada numericamente por uma série de Taylor com termos alternados. No entanto, para valores maiores de x
, ocorrem cancelamentos subtrativos significativos que afetam a acuracidade do resultado.

### a)
Escreva um programa que calcule o valor aproximado de $\sin(x)$ usando a série de Taylor:

$$\sin(x)\approx \sum^N_{n=0}(−1)^{n}\frac{x^{2n+1}}{(2n+1)!}$$

Implemente a função que aceita como entrada o valor de $x$ e o número de termos $N$.

### b)
Para $x\leq1$ ,compare o valor da série com o valor de $\sin(x)$ fornecido por uma biblioteca padrão (`math.sin`). Pare a soma automaticamente quando o valor do próximo termo da série for menor que $10^{−7}$ do somatório acumulado até aquele ponto. Imprima o valor final, o número de termos utilizados e o erro relativo.

```python
import math

def fatorial(n):
    """
    Calcula o fatorial de um número n.

    Args:
        n (int): O número para o qual o fatorial será calculado.

    Returns:
        int: O valor do fatorial de n.
    """
    if n < 0:
        raise ValueError("O fatorial não está definido para números negativos.")
    elif n == 0 or n == 1:
        return 1
    else:
        return n*fatorial(n - 1)

def sin_taylor_direto(x, N):
    """
    Calcula sin(x) usando a série de Taylor de forma direta.

    Args:
        x (float): O valor do ângulo em radianos.
        N (int): O número de termos a serem usados na série.

    Returns:
        float: O valor aproximado de sin(x).
    """
    soma_total = 0.0
    for n in range(0, N):
        # Termo do sinal: (-1)^(n)
        termo_sinal = (-1)**(n)
        
        # Termo da potência: x^(2n+1)
        expoente = 2 * n + 1
        termo_potencia = x**expoente
        
        # Termo do fatorial: (2n+1)!
        termo_fatorial = fatorial(expoente)
        
        # Termo completo da série
        termo_atual = termo_sinal * (termo_potencia / termo_fatorial)
        
        # Adiciona o termo atual à soma total
        soma_total += termo_atual
        
    return soma_total



# Valor de x em radianos (ex: 45 graus)
angulo_graus = float(input("Digite o ângulo em graus: "))
x_radianos = float(angulo_graus) * (math.pi / 180.0)  # Converte graus para radianos
x_radianos_math = math.radians(angulo_graus)
    
# Número de termos na série
num_termos = 85
    
print(f"Calculando sin({angulo_graus}°) ≈ sin({x_radianos:.4f} rad) com N = {num_termos} termos:\n")
    
# Calculando com a função direta
aprox_direta = sin_taylor_direto(x_radianos, num_termos)
print(f"  Aproximação (Direta):     {aprox_direta}")
    
# Valor real para comparação
valor_real = math.sin(x_radianos_math)
print(f"  Valor Real (math.sin):    {valor_real}")
    
# Verificando o erro da aproximação
erro = abs(valor_real - aprox_direta)
print(f"\n  Erro absoluto da aproximação eficiente: {erro:.2e}")
```

### d) (Exploração mais desafiadora)

Implemente a mesma aproximação da série de Taylor utilizando o tipo `decimal.Decimal` com precisão arbitrária. Faça o cálculo para $x=3\pi$ com 50 dígitos de precisão. Compare o valor obtido com o de `math.sin(3π)` e analise:

- O erro relativo usando ponto flutuante (`float`) e usando `decimal`.
- A diferença na quantidade de cancelamento observada.
- O custo computacional de usar precisão arbitrária.

```python
import math
import time

def fatorial(n):
    # (função fatorial como fornecida)
    if n < 0:
        raise ValueError("O fatorial não está definido para números negativos.")
    elif n == 0 or n == 1:
        return 1
    else:
        # Nota: Recursão pode ser lenta e atingir limites.
        return n * fatorial(n - 1)

def sin_taylor_float(x, N):
    soma_total = 0.0
    for n in range(0, N):
        termo_sinal = (-1)**(n)
        expoente = 2 * n + 1
        termo_potencia = x**expoente
        termo_fatorial = fatorial(expoente) # Usando a fatorial recursiva
        termo_atual = termo_sinal * (termo_potencia / termo_fatorial)
        soma_total += termo_atual
    return soma_total

# --- Execução com Float ---
x_radianos_float = 3 * math.pi
num_termos = 85

print(f"--- Calculando sin(3π) com Float (Padrão) ---")
start_time = time.time()
aprox_float = sin_taylor_float(x_radianos_float, num_termos)
end_time = time.time()
tempo_float = end_time - start_time

# O valor "exato" para float é o que a função math.sin calcula
valor_real_float = math.sin(x_radianos_float)

print(f"  Aproximação com Taylor (float): {aprox_float:e}")
print(f"  Valor de 'math.sin(3*pi)':      {valor_real_float:e}")
print(f"  Tempo de execução:              {tempo_float:.6f} segundos")
```

```python
from decimal import Decimal, getcontext
import time

def sin_taylor_decimal(x_decimal, N):
    """
    Calcula sin(x) usando a série de Taylor com precisão Decimal.
    """
    # Define a precisão para os cálculos. 50 + 3 para margem de segurança.
    getcontext().prec = 53

    if N == 0:
        return Decimal(0)
    
    # O primeiro termo da série (para n=0) é x.
    termo_atual = x_decimal
    soma_total = termo_atual
    
    # Calcula os termos subsequentes (de n=1 até N-1)
    for n in range(1, N):
        expoente_numerador = Decimal(2 * n + 1)
        expoente_denominador = Decimal(2 * n)
        
        # Fator multiplicativo para obter o próximo termo
        fator = - (x_decimal**2) / (expoente_numerador * expoente_denominador)
        
        termo_atual *= fator
        soma_total += termo_atual
        
    # Retorna o resultado com a precisão final desejada
    getcontext().prec = 50
    return +soma_total

# --- Execução com Decimal ---
# Define a precisão global
getcontext().prec = 53 # Usar precisão extra para definir Pi

# Criamos um valor de Pi com alta precisão
pi_decimal = Decimal("3.141592653589793238462643383279502884197169399375105820974944")
x_radianos_decimal = 3 * pi_decimal
num_termos = 85

print(f"\n--- Calculando sin(3π) com Decimal (Precisão de 50 dígitos) ---")
start_time = time.time()
aprox_decimal = sin_taylor_decimal(x_radianos_decimal, num_termos)
end_time = time.time()
tempo_decimal = end_time - start_time

# O valor exato de sin(3π).
valor_real_decimal = Decimal(math.sin(x_radianos_decimal))

print(f"  Aproximação com Taylor (Decimal): {aprox_decimal:e}")
print(f"  Valor de 'math.sin(3*pi)':        {valor_real_decimal}")
print(f"  Tempo de execução:                {tempo_decimal:.6f} segundos")
```
